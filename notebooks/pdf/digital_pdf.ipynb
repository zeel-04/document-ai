```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from pydantic import BaseModel

from doc_intelligence.llm import OpenAILLM
from doc_intelligence.processer import DocumentProcessor
```

```python
llm = OpenAILLM()
```

```python
processor = DocumentProcessor.from_digital_pdf(
    uri="../../data/Bank_eStmt_2025-07-14.pdf",
    llm=llm,
    include_line_numbers=True,
)
```

```python
class EndingBalance(BaseModel):
    ending_balance: float
    ending_balance_citation: processor.citation_type
    start_balance: float
    start_balance_citation: processor.citation_type


response = processor.extract(
    model="gpt-5-mini",
    reasoning={"effort": "low"},
    response_format=EndingBalance,
    page_numbers=[0, 1],
)
response.model_dump()
```
